```python
# returns an output to thrusters for one specific phase combo
def intcode_computer(puzzle_input, phase_combo):
    input = 0
    
    for phase in phase_combo:
        intcode = puzzle_input
        pointer = 0
        done = False
        use_phase = True

        while not done:
            # deteremine mode and opcode
            try:
                opcode = int(str(intcode[pointer])[-2:])
            except:
                opcode = int(str(intcode[pointer])[-1])
            try:
                mode_x = int(str(intcode[pointer])[-3])
            except:
                mode_x = 0
            try:
                mode_y = int(str(intcode[pointer])[-4])
            except:
                mode_y = 0

            # execute stuff based on opcode
            if opcode == 99:
                input = output
                done = True

            elif opcode == 1 or opcode == 2:
                if mode_x == 0:
                    x = intcode[intcode[pointer+1]]
                elif mode_x == 1:
                    x = intcode[pointer+1]
                if mode_y == 0:
                    y = intcode[intcode[pointer+2]]
                elif mode_y == 1:
                    y = intcode[pointer+2]
                
                if opcode == 1:
                    intcode[intcode[pointer+3]] = x + y
                else:
                    intcode[intcode[pointer+3]] = x * y
                pointer += 4

            elif opcode == 3:
#                 print('opcode 3')
                if use_phase:
                    intcode[intcode[pointer+1]] = phase
                    pointer += 2
                    use_phase = False
                else:
                    intcode[intcode[pointer+1]] = input
                    pointer += 2

            elif opcode == 4:
                output = intcode[intcode[pointer+1]]
                pointer += 2
#                 print(output)

            elif opcode == 5 or opcode == 6:
                if mode_x == 0:
                    x = intcode[intcode[pointer+1]]
                elif mode_x == 1:
                    x = intcode[pointer+1]
                if mode_y == 0:
                    y = intcode[intcode[pointer+2]]
                elif mode_y == 1:
                    y = intcode[pointer+2]
                
                if opcode == 5:
                    if x != 0:
                        pointer = y
                    else:
                        pointer += 3
                else:
                    if x == 0:
                        pointer = y
                    else:
                        pointer += 3

#             elif opcode == 7 or opcode == 8:
            else:
                if mode_x == 0:
                    x = intcode[intcode[pointer+1]]
                elif mode_x == 1:
                    x = intcode[pointer+1]
                if mode_y == 0:
                    y = intcode[intcode[pointer+2]]
                elif mode_y == 1:
                    y = intcode[pointer+2]
                
                if opcode == 7:
                    if x < y:
                        intcode[intcode[pointer+3]] = 1
                        pointer += 4
                    else:
                        intcode[intcode[pointer+3]] = 0
                        pointer += 4
                else:
                    if x == y:
                        intcode[intcode[pointer+3]] = 1
                        pointer += 4
                    else:
                        intcode[intcode[pointer+3]] = 0
                        pointer += 4

    return input
```

```python
# returns an output to thrusters for one specific phase combo
def intcode_computer_aaa(puzzle_input, phase_combo):
    input = 0
    amp_count = 0
    use_phase = True
    pointer = 0
    done = False
    intcode = puzzle_input
    phase_combo = list(phase_combo)
    print(intcode)

    while not done:
        print(f'starting while loop, pointer: {pointer}, amp_count: {amp_count}')
        # deteremine mode and opcode
        try:
            opcode = int(str(intcode[pointer])[-2:])
        except:
            opcode = int(str(intcode[pointer])[-1])
        try:
            mode_x = int(str(intcode[pointer])[-3])
        except:
            mode_x = 0
        try:
            mode_y = int(str(intcode[pointer])[-4])
        except:
            mode_y = 0

        # execute stuff based on opcode
        if opcode == 99:
            print('opcode 99')
            input = output
            done = True
            print('HALT!')

        elif opcode == 1 or opcode == 2:
            if mode_x == 0:
                x = intcode[intcode[pointer+1]]
            elif mode_x == 1:
                x = intcode[pointer+1]
            if mode_y == 0:
                y = intcode[intcode[pointer+2]]
            elif mode_y == 1:
                y = intcode[pointer+2]

            if opcode == 1:
                intcode[intcode[pointer+3]] = x + y
                print(f'opcode 1, add {x} to {y}, save it ({x+y}) to pos {intcode[pointer+3]}')
                print(intcode)
            else:
                intcode[intcode[pointer+3]] = x * y
                print(f'opcode 2, multiply {x} by {y}, save it ({x*y}) to pos {intcode[pointer+3]}')
                print(intcode)
            pointer += 4

        elif opcode == 3:
#             print('opcode 3')
            if use_phase and amp_count+5 in range(5,10):
                intcode[intcode[pointer+1]] = phase_combo[amp_count]
                pointer += 2
                use_phase = False
                print(f'opcode 3, use phase, save value {phase_combo[amp_count]} into pos {intcode[pointer+1]}')
                print(intcode)
            else:
                intcode[intcode[pointer+1]] = input
                pointer += 2
                print(f'opcode 3, use input ({input}), save value {input} into pos {intcode[pointer+1]}')
                print(intcode)

        elif opcode == 4:
            output = intcode[intcode[pointer+1]]
            print('opcode 4, give out output', output)
            print()
            pointer += 2
            amp_count += 1
            use_phase = True

        elif opcode == 5 or opcode == 6:
            if mode_x == 0:
                x = intcode[intcode[pointer+1]]
            elif mode_x == 1:
                x = intcode[pointer+1]
            if mode_y == 0:
                y = intcode[intcode[pointer+2]]
            elif mode_y == 1:
                y = intcode[pointer+2]

            if opcode == 5:
                if x != 0:
                    print(f'opcode 5, {x} != 0, so pointer = {y}')
                    pointer = y
                else:
                    print(f'opcode 5, {x} == 0, so do nothing')
                    pointer += 3
            else:
                print('opcode 6')
                if x == 0:
                    pointer = y
                else:
                    pointer += 3

        elif opcode == 7 or opcode == 8:
            if mode_x == 0:
                x = intcode[intcode[pointer+1]]
            elif mode_x == 1:
                x = intcode[pointer+1]
            if mode_y == 0:
                y = intcode[intcode[pointer+2]]
            elif mode_y == 1:
                y = intcode[pointer+2]

            if opcode == 7:
                print('opcode 7')
                if x < y:
                    intcode[intcode[pointer+3]] = 1
                    pointer += 4
                else:
                    intcode[intcode[pointer+3]] = 0
                    pointer += 4
            else:
                print('opcode 8')
                if x == y:
                    intcode[intcode[pointer+3]] = 1
                    pointer += 4
                else:
                    intcode[intcode[pointer+3]] = 0
                    pointer += 4

    return input
```
