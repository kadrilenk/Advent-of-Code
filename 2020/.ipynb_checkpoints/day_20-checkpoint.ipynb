```python
import re
import numpy as np
import itertools
import time
```

```python
input_data = [d.strip() for d in open('input20.txt')]
```

```python
tiles = {}

for i in [i for i,l in enumerate(input_data) if 'Tile' in l]:
    tiles[int(re.search('(\d+)',input_data[i]).group(1))] = input_data[i+1:i+11]
    
for t in tiles:
    for i, l in enumerate(tiles[t]):
        tiles[t][i] = [x for x in l]
    tiles[t] = np.array(tiles[t])
len(tiles)
```

```python
matching_pairs = []

for t in list(itertools.permutations([i for i in tiles],2)):
    for rot in itertools.product([0,1,2,3],[0,1,2,3]):
        tile1 = np.rot90(tiles[t[0]],rot[0])
        tile2 = np.rot90(tiles[t[1]],rot[1])
            
        if all(tile1[:,-1] == tile2[:,0]):
            matching_pairs.append([{t[0]:tile1},{t[1]:tile2}])
            
        if all(tile1.T[:,-1] == tile2.T[:,0]):
            matching_pairs.append([{t[0]:tile1.T},{t[1]:tile2.T}])
            
        if all(tile1.T[:,-1] == tile2[:,0]):
            matching_pairs.append([{t[0]:tile1.T},{t[1]:tile2}])
            
        if all(tile1[:,-1] == tile2.T[:,0]):
            matching_pairs.append([{t[0]:tile1},{t[1]:tile2.T}])

len(matching_pairs) # 1056
```

```python
matching_fours = []
for t in list(itertools.combinations(matching_pairs,2)):    
    tile1_id, tile2_id = [list(i.keys())[0] for i in t[0]]
    tile1, tile2 = [list(i.values())[0] for i in t[0]]
    tile3_id, tile4_id = [list(i.keys())[0] for i in t[1]]
    tile3, tile4 = [list(i.values())[0] for i in t[1]]
        
    if all(tile2[:,-1] == tile3[:,0]):
        matching_fours.append([{tile1_id:tile1},{tile2_id:tile2}, {tile3_id:tile3}, {tile4_id:tile4}])
len(matching_fours) # 950
```

```python
matching_eights = []
for t in list(itertools.combinations(matching_fours,2)):
    tile1_id, tile2_id, tile3_id, tile4_id = [list(i.keys())[0] for i in t[0]]
    tile1, tile2, tile3, tile4 = [list(i.values())[0] for i in t[0]]
    
    tile5_id, tile6_id, tile7_id, tile8_id = [list(i.keys())[0] for i in t[1]]
    tile5, tile6, tile7, tile8 = [list(i.values())[0] for i in t[1]]
        
    if all(tile4[:,-1] == tile5[:,0]):
        matching_eights.append([{tile1_id:tile1},{tile2_id:tile2},{tile3_id:tile3},{tile4_id:tile4},
                                {tile5_id:tile5},{tile6_id:tile6},{tile7_id:tile7},{tile8_id:tile8}])
len(matching_eights) # 750
```

```python
matching_rows = []
for t1 in matching_eights:
    for t2 in matching_eights:
        a = [list(i.values())[0] for i in t1[4:]]
        b = [list(i.values())[0] for i in t2[:4]]
        if all([all(a[i].flatten() == b[i].flatten()) for i in range(4)]):

            entry = []
            for i in range(8):
                entry.append({[list(i.keys())[0] for i in t1][i]:[list(i.values())[0] for i in t1][i]})
            for i in range(4):
                entry.append({[list(i.keys())[0] for i in t2[4:]][i]:[list(i.values())[0] for i in t2[4:]][i]})
            matching_rows.append(entry)

len(matching_rows)
# 474
```

```python
row_pairs = []
for row in list(itertools.combinations(matching_rows,2)): # 112101
    bottom = np.array([list(i.values())[0][-1,:] for i in row[0]]).flatten()
    top = np.array([list(i.values())[0][0,:] for i in row[1]]).flatten()
    if all(bottom==top):
        row_pairs.append([row[0],row[1]])
len(row_pairs) # 249
```

```python
row_fours = []
for row_pair in list(itertools.combinations(row_pairs,2)):
    
    r1_ids = [list(i.keys())[0] for i in row_pair[0][0]]
    r2_ids = [list(i.keys())[0] for i in row_pair[0][1]]
    r3_ids = [list(i.keys())[0] for i in row_pair[1][0]]
    r4_ids = [list(i.keys())[0] for i in row_pair[1][1]]
    
    r1 = np.array([list(i.values())[0] for i in row_pair[0][0]]).flatten()
    r2 = np.array([list(i.values())[0] for i in row_pair[0][1]]).flatten()
    r3 = np.array([list(i.values())[0] for i in row_pair[1][0]]).flatten()
    r4 = np.array([list(i.values())[0] for i in row_pair[1][1]]).flatten()
    
    if all(r2 == r3):
#         print('juhuu')
        print(r1_ids)
        print(r2_ids)
        print(r3_ids)
        print(r4_ids)
    print()
```
