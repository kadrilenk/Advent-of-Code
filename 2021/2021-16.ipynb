```python
import math
```

```python
with open('16.txt') as f:
    data = f.read().strip()
```

```python
def hex_to_binary(hexadecimal):
    decode = {'0':'0000', '1':'0001', '2':'0010', '3':'0011', '4':'0100', '5':'0101', '6':'0110', '7':'0111', 
              '8':'1000', '9':'1001', 'A':'1010', 'B':'1011', 'C':'1100', 'D':'1101', 'E':'1110', 'F':'1111'}
    binary = ''

    for h in hexadecimal:
        binary += decode[h]

    return binary
```

```python
def part_1(s):
    version = int(s[:3], 2)
    type_id = int(s[3:6], 2)

    if type_id != 4:
        length_type_id = s[6]

        if length_type_id == '0':
            total_length = int(s[7:7+15], 2)
            remainder = s[7+15:]
            if remainder.count('1') == 0:
                remainder = ''
            if remainder:
                version += part_1(remainder)

        else:
            sub_packets_count = int(s[7:7+11], 2)
            remainder = s[7+11:]
            if remainder.count('1') == 0:
                remainder = ''
            if remainder:
                version += part_1(remainder)

    else:
        literal = s[6:]
        i = 0
        number = ''

        while True:
            number += literal[i:i+5][1:]
            if literal[i:i+5][0] == '0':
                i += 5
                break
            i += 5

        remainder = literal[i:]
        if remainder.count('1') == 0:
            remainder = ''
        if remainder:
            version += part_1(remainder)

    return version
```

```python
print('Part 1:', part_1(hex_to_binary(data)))
```

```python
def part_2(s):
    version = int(s[:3], 2)
    type_id = int(s[3:6], 2)
    

    if type_id != 4:
        length_type_id = s[6]

        if length_type_id == '0':
            total_length = int(s[7:7+15], 2)
            l.append(f'type_{type_id}_start')
            
            bit_len = s[7+15:7+15+total_length]
            part_2(bit_len)
            l.append(f'type_{type_id}_stop')

            remainder = s[7+15+total_length:]
            if remainder.count('1') == 0:
                remainder = ''
            if remainder:
                part_2(remainder)

        else:
            sub_packets_count = int(s[7:7+11], 2)
            l.append(f'type_{type_id}__packets_{sub_packets_count}')

            remainder = s[7+11:]
            if remainder.count('1') == 0:
                remainder = ''
            if remainder:
                part_2(remainder)

    else:
        literal = s[6:]
        i = 0
        number = ''

        while True:
            number += literal[i:i+5][1:]
            if literal[i:i+5][0] == '0':
                i += 5
                break
            i += 5
        literal_value = int(number,2)
        l.append(literal_value)

        remainder = literal[i:]
        if remainder.count('1') == 0:
            remainder = ''
        if remainder:
            part_2(remainder)
```

```python
l = []
part_2(hex_to_binary(data))
```
