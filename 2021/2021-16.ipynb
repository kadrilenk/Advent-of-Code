```python
import math
```

```python
with open('16.txt') as f:
    input_data = f.read().strip()
```

```python
def hex_to_binary(hexadecimal):
    decode = {'0':'0000', '1':'0001', '2':'0010', '3':'0011', '4':'0100', '5':'0101', '6':'0110', '7':'0111', 
              '8':'1000', '9':'1001', 'A':'1010', 'B':'1011', 'C':'1100', 'D':'1101', 'E':'1110', 'F':'1111'}
    binary = ''

    for h in hexadecimal:
        binary += decode[h]

    return binary
```

```python
def part_1(s):
    version, type_id = int(s[:3], 2), int(s[3:6], 2)

    if type_id != 4:
        length_type_id = s[6]
        if length_type_id == '0':
            version += part_1(s[7+15:])
        else:
            version += part_1(s[7+11:])

    else:
        literal = s[6:]
        pos = min([i for i in range(0,len(literal),5) if literal[i]=='0'])+5
        remainder = literal[pos:]
        if remainder.count('1') == 0:
            remainder = ''
        if remainder:
            version += part_1(remainder)

    return version
```

```python
print('Part 1:', part_1(hex_to_binary(input_data)))
```

```python
def part_2(s,l):
    version, type_id = int(s[:3], 2), int(s[3:6], 2)

    if type_id != 4:
        length_type_id = s[6]

        if length_type_id == '0':
            total_length = int(s[7:7+15], 2)
            l.append(f'type_{type_id}_start')
            
            bit_len = s[7+15:7+15+total_length]
            part_2(bit_len,l)
            l.append(f'type_{type_id}_stop')

            remainder = s[7+15+total_length:]
            if remainder.count('1') == 0:
                remainder = ''
            if remainder:
                part_2(remainder,l)

        else:
            sub_packets_count = int(s[7:7+11], 2)
            l.append(f'type_{type_id}_packets_{sub_packets_count}')
            remainder = s[7+11:]
            part_2(remainder,l)

    else:
        literal = s[6:]
        pos = min([i for i in range(0,len(literal),5) if literal[i]=='0'])+5
        binary = ''.join([v for i,v in zip(list(range(5))*int(len(literal[:pos])/5),literal[:pos]) if i>0])
        literal_value = int(binary,2)
        l.append(literal_value)

        remainder = literal[pos:]
        if remainder.count('1') == 0:
            remainder = ''
        if remainder:
            part_2(remainder,l)
            
    return l
```

```python
def solve_part_2(l):
    while len(l) > 1:
        new_l = []
        i = 0

        while i < len(l):
            if str(l[i]).startswith('type_0_packets'):
                packets = int(l[i].replace('type_0_packets_',''))
                value_list = l[i+1:i+1+packets]
                if len([v for v in value_list if isinstance(v, int)])==len(value_list):
                    values = sum(value_list)
                    new_l.append(values)
                    i += len(value_list)+1
                else:
                    new_l.append(l[i])
                    i += 1

            elif str(l[i]).startswith('type_1_packets'):
                packets = int(l[i].replace('type_1_packets_',''))
                value_list = l[i+1:i+1+packets]
                if len([v for v in value_list if isinstance(v, int)])==len(value_list):
                    values = math.prod(value_list)
                    new_l.append(values)
                    i += len(value_list)+1
                else:
                    new_l.append(l[i])
                    i += 1

            elif str(l[i]).startswith('type_2_packets'):
                packets = int(l[i].replace('type_2_packets_',''))
                value_list = l[i+1:i+1+packets]
                if len([v for v in value_list if isinstance(v, int)])==len(value_list):
                    values = min(value_list)
                    new_l.append(values)
                    i += len(value_list)+1
                else:
                    new_l.append(l[i])
                    i += 1

            elif str(l[i]).startswith('type_3_packets'):
                packets = int(l[i].replace('type_3_packets_',''))
                value_list = l[i+1:i+1+packets]
                if len([v for v in value_list if isinstance(v, int)])==len(value_list):
                    values = max(value_list)
                    new_l.append(values)
                    i += len(value_list)+1
                else:
                    new_l.append(l[i])
                    i += 1

            elif str(l[i]).startswith('type_5_packets'):
                packets = int(l[i].replace('type_5_packets_',''))
                value_list = l[i+1:i+1+packets]
                if len([v for v in value_list if isinstance(v, int)])==len(value_list):
                    if value_list[0] > value_list[1]:
                        new_l.append(1)
                    else:
                        new_l.append(0)
                    i += len(value_list)+1
                else:
                    new_l.append(l[i])
                    i += 1

            elif str(l[i]).startswith('type_6_packets'):
                packets = int(l[i].replace('type_6_packets_',''))
                value_list = l[i+1:i+1+packets]
                if len([v for v in value_list if isinstance(v, int)])==len(value_list):
                    if value_list[0] < value_list[1]:
                        new_l.append(1)
                    else:
                        new_l.append(0)
                    i += len(value_list)+1
                else:
                    new_l.append(l[i])
                    i += 1

            elif str(l[i]).startswith('type_7_packets'):
                packets = int(l[i].replace('type_7_packets_',''))
                value_list = l[i+1:i+1+packets]
                if len([v for v in value_list if isinstance(v, int)])==len(value_list):
                    if value_list[0] == value_list[1]:
                        new_l.append(1)
                    else:
                        new_l.append(0)
                    i += len(value_list)+1
                else:
                    new_l.append(l[i])
                    i += 1

            elif l[i]=='type_0_start':
                stop = min([x for x,v in enumerate(l) if v=='type_0_stop' and x>i])
                value_list = l[i+1:stop]
                if len([v for v in value_list if isinstance(v, int)])==len(value_list):
                    values = sum(value_list)
                    new_l.append(values)
                    i += len(value_list)+2
                else:
                    new_l.append(l[i])
                    i += 1

            elif l[i]=='type_1_start':
                stop = min([x for x,v in enumerate(l) if v=='type_1_stop' and x>i])
                value_list = l[i+1:stop]
                if len([v for v in value_list if isinstance(v, int)])==len(value_list):
                    values = math.prod(value_list)
                    new_l.append(values)
                    i += len(value_list)+2
                else:
                    new_l.append(l[i])
                    i += 1

            elif l[i]=='type_2_start':
                stop = min([x for x,v in enumerate(l) if v=='type_2_stop' and x>i])
                value_list = l[i+1:stop]
                if len([v for v in value_list if isinstance(v, int)])==len(value_list):
                    values = min(value_list)
                    new_l.append(values)
                    i += len(value_list)+2
                else:
                    new_l.append(l[i])
                    i += 1

            elif l[i]=='type_3_start':
                stop = min([x for x,v in enumerate(l) if v=='type_3_stop' and x>i])
                value_list = l[i+1:stop]
                if len([v for v in value_list if isinstance(v, int)])==len(value_list):
                    values = max(value_list)
                    new_l.append(values)
                    i += len(value_list)+2
                else:
                    new_l.append(l[i])
                    i += 1

            elif l[i]=='type_5_start':
                stop = min([x for x,v in enumerate(l) if v=='type_5_stop' and x>i])
                value_list = l[i+1:stop]
                if len([v for v in value_list if isinstance(v, int)])==len(value_list):
                    if value_list[0] > value_list[1]:
                        new_l.append(1)
                    else:
                        new_l.append(0)
                    i += len(value_list)+2
                else:
                    new_l.append(l[i])
                    i += 1

            elif l[i]=='type_6_start':
                stop = min([x for x,v in enumerate(l) if v=='type_6_stop' and x>i])
                value_list = l[i+1:stop]
                if len([v for v in value_list if isinstance(v, int)])==len(value_list):
                    if value_list[0] < value_list[1]:
                        new_l.append(1)
                    else:
                        new_l.append(0)
                    i += len(value_list)+2
                else:
                    new_l.append(l[i])
                    i += 1

            elif l[i]=='type_7_start':
                stop = min([x for x,v in enumerate(l) if v=='type_7_stop' and x>i])
                value_list = l[i+1:stop]
                if len([v for v in value_list if isinstance(v, int)])==len(value_list):
                    if value_list[0] == value_list[1]:
                        new_l.append(1)
                    else:
                        new_l.append(0)
                    i += len(value_list)+2
                else:
                    new_l.append(l[i])
                    i += 1

            else:
                new_l.append(l[i])
                i += 1

        l = new_l

    return l[0]
```
